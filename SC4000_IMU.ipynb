```python
import os
from dotenv import load_dotenv
import pandas as pd
import glob
```

```python
train_deviceimu_path = "../train/2020-07-24-US-MTV-1/GooglePixel4/device_imu.csv"
train_deviceimu = pd.read_csv(train_deviceimu_path)

print(train_deviceimu.head())
```

```python
parent_folder = "../train"

# retrieve device.imu files from train
imu_files = glob.glob(os.path.join(parent_folder, "*", "*", "device_imu.csv"))
print(imu_files) 
```

```python
imu_list = []

for file in imu_files:
    imu_file = pd.read_csv(file)
    parts = file.split(os.sep) # eg ['train', '2021-12-28-US-MTV-1', 'GooglePixel5', 'device_imu.csv']
    imu_file["drive_id"] = parts[-3] 
    imu_file["device"] = parts[-2]
    imu_list.append(imu_file)

# combine all
imu_df = pd.concat(imu_list, ignore_index=True)
print(imu_df.head())
print()
print(imu_df["device"].unique())
print()
print(imu_df["drive_id"].unique())
```

```python
imu_acc_df = imu_df[imu_df["MessageType"] == "UncalAccel"][["utcTimeMillis", "MeasurementX", "MeasurementY", "MeasurementZ", "drive_id", "device"]]

print(imu_acc_df.head())
```

```python
imu_mag_df = pd.DataFrame()

imu_uncalmag_df = imu_df[imu_df["MessageType"] == "UncalMag"][["utcTimeMillis", "MeasurementX", "MeasurementY", "MeasurementZ", "BiasX", "BiasY", "BiasZ", "drive_id", "device"]]
imu_mag_df["utcTimeMillis"] = imu_uncalmag_df["utcTimeMillis"]
imu_mag_df["MeasurementX"] = imu_uncalmag_df["MeasurementX"] + imu_uncalmag_df["BiasX"]
imu_mag_df["MeasurementY"] = imu_uncalmag_df["MeasurementY"] + imu_uncalmag_df["BiasY"]
imu_mag_df["MeasurementZ"] = imu_uncalmag_df["MeasurementZ"] + imu_uncalmag_df["BiasZ"]
imu_mag_df["drive_id"] = imu_uncalmag_df["drive_id"]
imu_mag_df["device"] = imu_uncalmag_df["device"]

print(imu_mag_df.head())
```

```python
imu_gyro_df = imu_df[imu_df["MessageType"] == "UncalGyro"][["utcTimeMillis", "MeasurementX", "MeasurementY", "MeasurementZ", "drive_id", "device"]]

print(imu_gyro_df.head())
```
